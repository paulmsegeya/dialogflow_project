```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import random
```

```python
r = requests.get('https://www.facebook.com/business/help/169249477193317')
```

```python
# r.text
```

```python
r.encoding
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
a_tags = soup.find_all('a', class_='_3cr5 _5j3- _53m5 _1b60 _7p5k _1s6c')
```

```python
a_tags[0].text
```

```python
hrefs = [a_tag.get('href') for a_tag in a_tags]
```

```python
def parse_qna(soup):
    h2 = soup.find('h2', class_='_2s4h')
    div = soup.find('div', class_='_5pfd')
    if h2 is not None and div is not None:
        return{
            'title': h2.text,
            'content': div.text
        }
```

```python
data = []
for href in hrefs:
    url = 'https://www.facebook.com' + href
    print('Requesting {}'.format(url))
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    result = parse_qna(soup)
    data.append(result)
    print('Result')
    print(result)
    
    time.sleep(0.5 + random.random())
```

```python
df = pd.DataFrame([d for d in data if d is not None])
```
