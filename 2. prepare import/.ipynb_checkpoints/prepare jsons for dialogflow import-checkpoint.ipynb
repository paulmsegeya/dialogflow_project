```python
import pandas as pd
import os
import json
```

```python
df = pd.read_csv('facebook_ad_qna.csv')
df
```

```python
agent_name = 'Facebook Ad Q&A'
```

```python
intents = []

for i, row in df.iterrows():
    title = row['title']
    content = row['content']
    
    qna = {}
    qna['{}_usersays_zh-hk.json'.format(title)] = [
      {
#         "id": "08c57e8f-03f6-44ef-a23b-d914d639abe7",
        "data": [
          {
            "text": title.replace('關於', '咩係'),
            "userDefined": False
          }
        ],
        "isTemplate": False,
        "count": 0,
        "updated": 0
      },
    ]
    
    qna['{}.json'.format(title)] = {
#       "id": "38438e5b-2313-40d0-9a1f-7b3662266e8c",
      "name": title,
      "auto": True,
      "contexts": [],
      "responses": [
        {
          "resetContexts": False,
          "affectedContexts": [],
          "parameters": [],
          "messages": [
            {
              "type": 0,
              "lang": "zh-hk",
              "speech": content
            }
          ],
          "defaultResponsePlatforms": {},
          "speech": []
        }
      ],
      "priority": 500000,
      "webhookUsed": False,
      "webhookForSlotFilling": False,
      "fallbackIntent": False,
      "events": []
    }
    
    intents.append(qna)
```

```python
if not os.path.exists('intents'):
    os.rmdir('intents')
os.makedirs('intents')
```

```python
for jsons in intents:
    for key in jsons.keys():
        intents_dir = path = os.path.join(os.getcwd(), 'intents')
        filename = key

        with open(os.path.join(intents_dir, filename), 'w') as f:
            json.dump(jsons[key], f)
```
